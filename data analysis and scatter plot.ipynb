```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('airbnb_bologna_0310_2016-02-07.csv')
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['overall_satisfaction'].fillna(df['overall_satisfaction'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(5, 7))
plt.scatter(df.overall_satisfaction, df.latitude, alpha=0.2)
plt.title('Scatter Plot of Satisfaction vs latitude')
plt.xlabel('overall_satisfaction')
plt.ylabel('latitude')
plt.show()
```

```python
plt.figure(figsize=(5, 4))
plt.scatter(df.overall_satisfaction, df.price, alpha=0.2)
plt.title('Scatter Plot of Satisfaction vs price')
plt.xlabel('overall_satisfaction')
plt.ylabel('price')
plt.show()
```
